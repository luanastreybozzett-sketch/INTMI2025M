Peça um número inteiro e informe se ele é par. Crie um algoritmo em Python que verifique se o número é divisível por 2.

```python
# Pede ao usuário para digitar um número inteiro
numero_str = input("Digite um número inteiro: ")

try:
    # Converte a entrada do usuário para um número inteiro
    numero = int(numero_str)

    # Verifica se o número é divisível por 2 (resto 0)
    if numero % 2 == 0:
        print(f"O número {numero} é par.")
    else:
        print(f"O número {numero} é ímpar.")

except ValueError:
    # Se a conversão falhar, informa que a entrada não é um número inteiro válido
    print("Entrada inválida. Por favor, digite um número inteiro.")
```

Peça a quantidade de horas trabalhadas e o valor por hora. Se trabalhou até 40 horas, calcule o salário normalmente (horas × valor por hora). Se trabalhou mais de 40 horas, as horas extras devem ter um adicional de 50% (horas extras × valor por hora × 1.5). Crie um algoritmo em Python que realize esse cálculo corretamente.

```python
# Programa para calcular o salário com base nas horas trabalhadas

# Entrada: pedir os dados ao usuário
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_por_hora = float(input("Digite o valor pago por hora (R$): "))

# Processamento
if horas_trabalhadas <= 40:
    salario = horas_trabalhadas * valor_por_hora
else:
    horas_extras = horas_trabalhadas - 40
    salario = (40 * valor_por_hora) + (horas_extras * valor_por_hora * 1.5)

# Saída
print(f"Seu salário é: R${salario:.2f}")
```

O objetivo é criar um programa Python que simule uma calculadora interativa, capaz de realizar diferentes tipos de análises e conversões, e que só termine quando o usuário explicitamente decidir sair.

```python
import time

def multiplos_de_5(inicio, fim):
    """
    Exibe os múltiplos de 5 dentro de um intervalo [inicio, fim] e conta quantos são.
    """
    if inicio > fim:
        print(" Erro: O valor inicial deve ser menor ou igual ao valor final.")
        return

    print(f"\n Múltiplos de 5 no intervalo de {inicio} a {fim}:")
    multiplos = []
    for num in range(inicio, fim + 1):
        if num % 5 == 0:
            multiplos.append(num)

    if multiplos:
        print(", ".join(map(str, multiplos)))
        print(f" Total de múltiplos de 5 encontrados: {len(multiplos)}")
    else:
        print("Nenhum múltiplo de 5 encontrado neste intervalo.")

def contagem_pares(inicio, fim):
    """
    Conta e exibe os números pares dentro de um intervalo [inicio, fim].
    """
    if inicio > fim:
        print(" Erro: O valor inicial deve ser menor ou igual ao valor final.")
        return

    print(f"\n Números pares no intervalo de {inicio} a {fim}:")
    pares = []
    for num in range(inicio, fim + 1):
        if num % 2 == 0:
            pares.append(num)

    if pares:
        print(", ".join(map(str, pares)))
        print(f" Total de números pares encontrados: {len(pares)}")
    else:
        print("Nenhum número par encontrado neste intervalo.")

def verificador_primo(numero):
    """
    Verifica se um número inteiro positivo é primo.
    """
    if numero < 2:
        print(f"\n O número {numero} não é primo (primos são maiores ou iguais a 2).")
        return

    # Um número é primo se for divisível apenas por 1 e por ele mesmo.
    # Podemos otimizar checando divisibilidade apenas até a raiz quadrada do número.
    # Para simplicidade, checaremos até (numero // 2)
    for i in range(2, int(numero**0.5) + 1):
        if numero % i == 0:
            print(f"\n O número {numero} **NÃO** é primo.")
            print(f"Ele é divisível por {i}.")
            return

    print(f"\n O número {numero} **É PRIMO**!")

def conversor_temperatura(celsius):
    """
    Converte uma temperatura de Celsius (Cº) para Fahrenheit (Fº).
    Fórmula: F = C * (9/5) + 32
    """
    fahrenheit = (celsius * 9/5) + 32
    print(f"\n A temperatura de {celsius:.2f}ºC é equivalente a **{fahrenheit:.2f}ºF**.")


def exibir_menu():
    """
    Exibe o menu principal da calculadora.
    """
    print("\n" + "="*35)
    print("**CALCULADORA INTERATIVA**".center(35))
    print("="*35)
    print("Digite o número da opção desejada:")
    print("1 - Múltiplos de 5 em um Intervalo")
    print("2 - Contagem de Números Pares em um Intervalo")
    print("3 - Verificador de Número Primo")
    print("4 - Conversor de Temperatura (Cº para Fº)")
    print("0 - Sair do Programa")
    print("-" * 35)

def main():
    """
    Função principal que contém o loop da calculadora.
    """
    while True:
        exibir_menu()

        try:
            opcao = input("Opção: ").strip()

            if opcao == '0':
                print("\n **Saindo do Programa... Obrigado por usar a Calculadora Interativa!**")
                break # Interrompe o loop principal

            elif opcao == '1':
                print("\n--- [ Opção 1: Múltiplos de 5 ] ---")
                try:
                    inicio = int(input("Digite o valor inicial do intervalo: "))
                    fim = int(input("Digite o valor final do intervalo: "))
                    multiplos_de_5(inicio, fim)
                except ValueError:
                    print(" Entrada inválida. Por favor, digite apenas números inteiros para os limites do intervalo.")

            elif opcao == '2':
                print("\n--- [ Opção 2: Contagem de Pares ] ---")
                try:
                    inicio = int(input("Digite o valor inicial do intervalo: "))
                    fim = int(input("Digite o valor final do intervalo: "))
                    contagem_pares(inicio, fim)
                except ValueError:
                    print(" Entrada inválida. Por favor, digite apenas números inteiros para os limites do intervalo.")

            elif opcao == '3':
                print("\n--- [ Opção 3: Verificador de Primo ] ---")
                try:
                    numero = int(input("Digite o número inteiro a ser verificado: "))
                    verificador_primo(numero)
                except ValueError:
                    print(" Entrada inválida. Por favor, digite apenas um número inteiro.")

            elif opcao == '4':
                print("\n--- [ Opção 4: Conversor Cº para Fº ] ---")
                try:
                    celsius = float(input("Digite a temperatura em Celsius (Cº): "))
                    conversor_temperatura(celsius)
                except ValueError:
                    print(" Entrada inválida. Por favor, digite um valor numérico para a temperatura.")

            else:
                print("\n Opção inválida. Por favor, escolha um número de 0 a 4.")

            # Espera um pouco antes de exibir o menu novamente para melhor leitura
            if opcao != '0':
                time.sleep(2)
                input("\nPressione ENTER para continuar...") # Pausa para o usuário ler o resultado

        except KeyboardInterrupt:
            # Captura Ctrl+C para sair de forma limpa, caso o usuário não use o '0'
            print("\n\n **Programa interrompido. Saindo!**")
            break
        except Exception as e:
            # Captura outras exceções inesperadas
            print(f"\n Ocorreu um erro inesperado: {e}")
            time.sleep(2)


# Garante que a função 'main' seja executada quando o script for iniciado
if __name__ == "__main__":
    main()
```
